```python
#send get request to the webpage server to get the source code of the page

page = requests.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
page
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
tables=soup.find_all('td')[4:184]
```

```python
table=[]
for i in tables:
    table.append(i.text)
```

```python
table
```

```python
#now slicing for required data

rank=[]
for i in table[::6]:
    rank.append(i)

name=[]
for i in table[1::6]:
    name.append(i[:i.index('[')])
    
artist=[]
for i in table[2::6]:
    artist.append(i)
    
views=[]
for i in table[3::6]:
    views.append(i)
    
date=[]
for i in table[4::6]:
    date.append(i)
```

```python
#checking for length of the lists

print(len(rank),len(name),len(artist),len(views),len(date))
```

```python
#finally forming a dataframe for the extracted data

df=pd.DataFrame({'Rank':rank,'Title':name,'Artist':artist,'Views(in billions)':views,'Upload date':date})
df.style.set_caption('Most viewed videos on youtube')
df.set_index('Rank')
```

```python
#loading the website on driver

driver.get('https://www.imdb.com/list/ls095964455/')

#requesting webpage server for source code

page=requests.get('https://www.imdb.com/list/ls095964455/')
print(page)

#loading the content

soup=BeautifulSoup(page.content,'html.parser')
```

```python
#scraping name

name = soup.find_all("h3", class_="lister-item-header")
# get text from series name web elements
series_name = [] #empty list
for i in name:
    for j in i.find_all("a"):
        series_name.append(j.text)


# Year of release
year = soup.find_all("span",class_="lister-item-year text-muted unbold")
year_span = [] #empty list
for k in year:
    a=k.text.replace('(','')
    year_span.append(a.replace(')','')) 


      
# IMDB Rating
rating = soup.find_all("div",class_="ipl-rating-star small")

# scrape text from rating web element
IMDB_rating = [] #empty list
for i in rating:
      IMDB_rating.append(float(i.text))
        
#run time 

run=[]
for i in soup.find_all('span',class_='runtime'):
    run.append(i.text)

#genre

genre=[]
for i in soup.find_all('span',class_='genre'):
    genre.append(i.text.replace('\n','').strip())
    
#votes

vote=[]
for i in driver.find_elements_by_name("nv"):
    vote.append(i.text)
```

```python
#checking the length of the lists formed

print(len(series_name),len(year_span),len(rating),len(run),len(genre),len(vote))
```

```python
#finally forming a dataframe for the extracted data

df=pd.DataFrame({'Name':series_name,'Year span':year_span,'Run-time':run,'Genre':genre,'Rating':(rating),'Votes':vote})
df.style.set_caption('TOP 100 MOST WATCHED TV SHOWS OF ALL TIME')
```
