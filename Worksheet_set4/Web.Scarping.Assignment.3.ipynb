```python
#search bar 

driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys(str(input('Enter smartphone name/brand: ')))
```

```python
#search button

driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()

#scraping name description

productdescription=[]
for i in driver.find_elements_by_xpath("//div[@class='_4rR01T']"):
    productdescription.append(i.text)
    
#slicing for name

pname=[]
for i in productdescription:
    i=(i[:i.index('(')]).strip()
    pname.append(i)
    
#slicing for color

color=[]
for i in productdescription:
    i=i[i.index('(')+1:i.index(',')]
    color.append(i)

#slicing for ROM

ROM=[]
for i in productdescription:
    i=i[i.index(',')+2:-1]
    ROM.append(i)   

#scraping product's features description

description=[]
for i in driver.find_elements_by_xpath("//li[@class='rgWa7D']"):
    description.append(i.text)
    
#slicing for ROM

RAM=[]
for i in description[::5]:
    i=(i[:9]).strip()
    RAM.append(i)
    
#slicing for display

Display=[]
for i in description[1::5]:
    Display.append(i)
    
#slicing for rear camera

rcamera=[]
for i in description[2::5]:
    rcamera.append(i)
    
#slicing for battery

battery=[]
for i in description[3::5]:
    battery.append(i)
    
#slicing for price

price=[]
for i in driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']"):
    price.append(i.text)
    
#slicing for url

url=[]
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    url.append(i.get_attribute('href'))
```

```python
#checking the length of the lists created

print(len(pname),len(color),len(RAM),len(ROM),len(Display),len(rcamera),len(battery),len(price),len(url))
```

```python
#finally creating dataframe for the extracted data

import pandas as pd

df=pd.DataFrame({'Product Name':pname,'Color':color,'RAM':RAM,'ROM':ROM,'Display':Display,'Rear Camera':rcamera,'Battery':battery,'Price':price,'URL':url})
```

```python
df
```

```python
#saving file to csv

df.to_csv('smartphones.csv')
```

```python
#lets first connect to the webdriver

driver=webdriver.Chrome('chromedriver.exe')

#opening up hostelworld.com website on automated chrome window

driver.get('https://www.hostelworld.com/')

#entering location in the search bar

driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input').send_keys('London')

#clicking on search button

driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]').click()

driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[5]/button').click()

#scraping name

name=[]
for i in driver.find_elements_by_xpath("//h2[@class='title title-6']"):
    name.append(i.text)
    
#scraping description

description=[]
for i in driver.find_elements_by_xpath("//span[@class='description']"):
    description.append(i.text)
    
#now slicing for propert description and distance from centre

propertyy=[]
for i in description:
    i=i[:i.index('-')-1]
    propertyy.append(i)
    
distance=[]
for i in description:
    i=i[i.index('-')+2:]
    distance.append(i)
    
#scraping rating summary

summ=[]
for i in driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']"):
    summ.append(i.text.replace('\n',','))
    
#slicing for ratings

rating=[]
for i in summ:
    i=i[:i.index(',')]
    rating.append(i)

#scraping for over all reviews

over=[]
for i in driver.find_elements_by_xpath("//div[@class='keyword']")[2:]:
    over.append(i.text)
    
#scraping total reviews

total=[]
for i in driver.find_elements_by_xpath("//div[@class='reviews']"):
    total.append(i.text)
    
#scraping price column

price=[]
for i in driver.find_elements_by_xpath("//div[@class='price-col']"):
    price.append(i.text.replace('\n','-'))
    
#now slicing for private prices and dorm prices respectively

pvt=[]
for i in price[::2]:
    pvt.append(i)
    
dorm=[]
for i in price[1::2]:
    dorm.append(i)
    
#scraping facilities

fac=[]
for i in driver.find_elements_by_xpath("//div[@class='facilities-label facilities']"):
    fac.append(i.text.replace('\n',','))
```

```python
#checking the length of the lists formed

print(len(name),len(distance),len(rating),len(total),len(over),len(pvt),len(dorm),len(fac),len(propertyy))
```

```python
#finally forming a dataframe for the extracted data

import pandas as pd

df=pd.DataFrame({'NAME':name,'DISTANCE':distance,'RATING':rating,'TOTAL REVIEWS':total,'OVER ALL REVIEW':over,'PRIVATES FROM PRICES':pvt,'DORMS FROM PRICES':dorm,'FACILITIES':fac,'PROPERTY TYPE':propertyy})
df
```
