```python
company_names
```

```python
for i in loc:
    locations.append(i.text)
    
locations=locations[0:10]
```

```python
#extracting experience tags

exp=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
exp[0:4]
```

```python
for i in exp:
    experience.append(i.text)

experience=experience[0:10]
```

```python
#now checking the lengths of the lists created

print(len(job_titles),len(company_names),len(locations),len(experience))
```

```python
#finally forming a dataframe for the extracted data

df=pd.DataFrame({'JOB TITLE':job_titles,'COMPANY':company_names,'LOCATION':locations,'EXPERIENCE REQUIRED':experience})
df.style.set_caption('DATA ANALYST JOBS IN BENGALORE')
```

```python
#closing the driver

driver.close()
```
